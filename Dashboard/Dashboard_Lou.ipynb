## Import Data

```python
file_path = Path('coin_Bitcoin.csv')
bitcoin_df = pd.read_csv(file_path,parse_dates=True, infer_datetime_format=True, #index_col = 'Date'
                        )

file_path_2 = Path('coin_Ethereum.csv')
eth_df = pd.read_csv(file_path_2,parse_dates=True, infer_datetime_format=True, #index_col = 'Date'
                    )

file_path_3 = Path('coin_Litecoin.csv',parse_dates=True, infer_datetime_format=True, #index_col = 'Date'
               )
lite_df = pd.read_csv(file_path_3)

file_path_4 = Path('coin_XRP.csv',parse_dates=True, infer_datetime_format=True, #index_col = 'Date'
                  )
xrp_df = pd.read_csv(file_path_4)

file_path_5 = Path('Bitcoin_mdy.csv')
bitcoin_mdy_df = pd.read_csv(file_path_5)

file_path_6 = Path('Ethereum_mdy.csv')
eth_mdy_df = pd.read_csv(file_path_6)

file_path_7 = Path('Litecoin_mdy.csv')
lite_mdy_df = pd.read_csv(file_path_7)

file_path_8 = Path('XRP_mdy.csv')
xrp_mdy_df = pd.read_csv(file_path_8)
```

# Panel Visualizations 

```python
def btc_price_vol():
    """Bitcoin Price/ Volume"""
    BTC_P_V = pd.concat([bitcoin_df], axis = 0, join = 'inner')
    start_date = pd.to_datetime('2016-01-01')
    end_date = pd.to_datetime('2021-02-27')                         
    BTC_P_V['Date'] = pd.to_datetime(BTC_P_V['Date']) 
    new_df = (BTC_P_V['Date']>= start_date) & (BTC_P_V['Date']<= end_date)
    df1 = BTC_P_V.loc[new_df]
    stock_data = df1.set_index('Date')
    top_plt = plt.subplot2grid((5,4), (0, 0), rowspan=3, colspan=4)
    top_plt.plot(stock_data.index, stock_data["Close"])
    plt.title('Historical prices of Bitcoin [01-01-2016 to 01-27-2021]')
    bottom_plt = plt.subplot2grid((5,4), (3,0), rowspan=1, colspan=4)
    bottom_plt.bar(stock_data.index, stock_data['Volume'])
    plt.title('\nBitcoin Trading Volume', y=-0.60)
    plt.gcf().set_size_inches(12,8)
```

```python
def volume_per_month ():
    """Average Monthly Volume"""
    MDY = pd.concat([bitcoin_mdy_df['Volume'], eth_mdy_df['Volume'], lite_mdy_df['Volume'], xrp_mdy_df['Volume'], bitcoin_mdy_df['Close'], eth_mdy_df['Close'], lite_mdy_df['Close'], xrp_mdy_df['Close'], bitcoin_mdy_df['Marketcap'], eth_mdy_df['Marketcap'], lite_mdy_df['Marketcap'], xrp_mdy_df['Marketcap'], bitcoin_mdy_df['Year'], bitcoin_mdy_df['Month'], bitcoin_mdy_df['Day']], axis =1, join = 'inner')
    MDY.columns = ['BTC Vol', 'ETH Vol', 'LTC Vol', 'XRP Vol', 'BTC Close', 'ETH Close', 'LTC Close', 'XRP Close', 'BTC Cap', 'ETH Cap', 'LTC Cap', 'XRP Cap', 'Year', 'Month', 'Day']
    a_1 = MDY.groupby('Month').mean()
    return a_1.hvplot.line(
        y = ('BTC Vol', 'ETH Vol', 'LTC Vol', 'XRP Vol'),
        title = 'Average Volume By Month').opts(yformatter = '%.0f')
```

```python
def price_per_month ():
    """Average Monthly Price"""
    MDY = pd.concat([bitcoin_mdy_df['Volume'], eth_mdy_df['Volume'], lite_mdy_df['Volume'], xrp_mdy_df['Volume'], bitcoin_mdy_df['Close'], eth_mdy_df['Close'], lite_mdy_df['Close'], xrp_mdy_df['Close'], bitcoin_mdy_df['Marketcap'], eth_mdy_df['Marketcap'], lite_mdy_df['Marketcap'], xrp_mdy_df['Marketcap'], bitcoin_mdy_df['Year'], bitcoin_mdy_df['Month'], bitcoin_mdy_df['Day']], axis =1, join = 'inner')
    MDY.columns = ['BTC Vol', 'ETH Vol', 'LTC Vol', 'XRP Vol', 'BTC Close', 'ETH Close', 'LTC Close', 'XRP Close', 'BTC Cap', 'ETH Cap', 'LTC Cap', 'XRP Cap', 'Year', 'Month', 'Day']
    a_2 = MDY.groupby('Month').mean()
    return a_2.hvplot.line(
        y = ('BTC Close', 'ETH Close', 'LTC Close', 'XRP Close'),
        title = ' Average Price By Month',
        ylim = (-1000, 15000)).opts(yformatter = '%.0f') 
```

```python
def basic_corr ():
    """Price and Volume Correlation"""
    MDY = pd.concat([bitcoin_mdy_df['Volume'], eth_mdy_df['Volume'], lite_mdy_df['Volume'], xrp_mdy_df['Volume'], bitcoin_mdy_df['Close'], eth_mdy_df['Close'], lite_mdy_df['Close'], xrp_mdy_df['Close'], bitcoin_mdy_df['Marketcap'], eth_mdy_df['Marketcap'], lite_mdy_df['Marketcap'], xrp_mdy_df['Marketcap'], bitcoin_mdy_df['Year'], bitcoin_mdy_df['Month'], bitcoin_mdy_df['Day']], axis =1, join = 'inner')
    MDY.columns = ['BTC Vol', 'ETH Vol', 'LTC Vol', 'XRP Vol', 'BTC Close', 'ETH Close', 'LTC Close', 'XRP Close', 'BTC Cap', 'ETH Cap', 'LTC Cap', 'XRP Cap', 'Year', 'Month', 'Day']
    corr_df = pd.concat([MDY])
    basic_corr_df = pd.concat([MDY])
    del basic_corr_df['Month'], basic_corr_df['Day'], basic_corr_df['Year'], basic_corr_df['BTC Cap'], basic_corr_df['ETH Cap'], basic_corr_df['LTC Cap'], basic_corr_df['XRP Cap']
    basic_corr = basic_corr_df.corr()
    plt.subplots(figsize=(10,5))
    sns.heatmap(basic_corr, vmin = -1, vmax = 1, annot = True)
```

```python
def big_corr():
    """Price to Previous Day Volume Correlation"""
    MDY = pd.concat([bitcoin_mdy_df['Volume'], eth_mdy_df['Volume'], lite_mdy_df['Volume'], xrp_mdy_df['Volume'], bitcoin_mdy_df['Close'], eth_mdy_df['Close'], lite_mdy_df['Close'], xrp_mdy_df['Close'], bitcoin_mdy_df['Marketcap'], eth_mdy_df['Marketcap'], lite_mdy_df['Marketcap'], xrp_mdy_df['Marketcap'], bitcoin_mdy_df['Year'], bitcoin_mdy_df['Month'], bitcoin_mdy_df['Day']], axis =1, join = 'inner')
    MDY.columns = ['BTC Vol', 'ETH Vol', 'LTC Vol', 'XRP Vol', 'BTC Close', 'ETH Close', 'LTC Close', 'XRP Close', 'BTC Cap', 'ETH Cap', 'LTC Cap', 'XRP Cap', 'Year', 'Month', 'Day']
    corr_df = pd.concat([MDY])
    del corr_df['Month'], corr_df['Day'], corr_df['Year'], corr_df['BTC Cap'], corr_df['ETH Cap'], corr_df['LTC Cap'], corr_df['XRP Cap']
    corr_df['BTC Vol -1'] = corr_df['BTC Vol'].shift(-1)
    corr_df['ETH Vol -1'] = corr_df['ETH Vol'].shift(-1)
    corr_df['LTC Vol -1'] = corr_df['LTC Vol'].shift(-1)
    corr_df['XRP Vol -1'] = corr_df['XRP Vol'].shift(-1)
    corr_df['BTC Vol -2'] = corr_df['BTC Vol'].shift(-2)
    corr_df['ETH Vol -2'] = corr_df['ETH Vol'].shift(-2)
    corr_df['LTC Vol -2'] = corr_df['LTC Vol'].shift(-2)
    corr_df['XRP Vol -2'] = corr_df['XRP Vol'].shift(-2)
    corr_df['BTC Vol -3'] = corr_df['BTC Vol'].shift(-3)
    corr_df['ETH Vol -3'] = corr_df['ETH Vol'].shift(-3)
    corr_df['LTC Vol -3'] = corr_df['LTC Vol'].shift(-3)
    corr_df['XRP Vol -3'] = corr_df['XRP Vol'].shift(-3)
    corr_df['BTC Vol -4'] = corr_df['BTC Vol'].shift(-4)
    corr_df['ETH Vol -4'] = corr_df['ETH Vol'].shift(-4)
    corr_df['LTC Vol -4'] = corr_df['LTC Vol'].shift(-4)
    corr_df['XRP Vol -4'] = corr_df['XRP Vol'].shift(-4)
    corr_df['BTC Vol -5'] = corr_df['BTC Vol'].shift(-5)
    corr_df['ETH Vol -5'] = corr_df['ETH Vol'].shift(-5)
    corr_df['LTC Vol -5'] = corr_df['LTC Vol'].shift(-5)
    corr_df['XRP Vol -5'] = corr_df['XRP Vol'].shift(-5)
    big_corr = corr_df.corr()
    plt.subplots(figsize=(20,10))
    sns.heatmap(big_corr, vmin = -1, vmax = 1, annot = True)
```

```python
 # Create a Title for the Dashboard
home_page = pn.Column('How does volume relate to crypto currency and what can it tell us about price changes?')
price_by_volume = pn.Column('Price by Volume for Crypto Currencies')
                           
pbv = pn.Column(btc_price_vol(),
                eth_price_vol(),
                ltc_price_vol(),
                xrp_price_vol())
correlation_maps = pn.Column(basic_corr(),
                             big_corr())
price_vol_month = pn.Column(volume_per_month(),
                            price_per_month())

my_dashboard = pn.Tabs(
    ('Welcome', home_page),
    ('Price By Volume', pbv),
    ('Price to Volume Correlation', correlation_maps),
    ('Price and Volume by Month', price_vol_month))
```
